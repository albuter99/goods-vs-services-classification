```python
import pandas as pd

df = pd.read_csv("amazon.csv")
df.head()
```

```python
df.info()

```

```python
cols_to_drop = [
    "product_id",
    "discounted_price",
    "actual_price",
    "discount_percentage",
    "rating",
    "rating_count",
    "user_id",
    "user_name",
    "img_link",
    "product_link",
    "review_id"
]

df_clean = df.drop(columns=cols_to_drop)

df_clean.head()
```

```python
df_clean["goods_or_services"] = "goods"
df_clean.head()
df_clean["goods_or_services"].value_counts()
```

```python
df_clean.head()
```

```python
df_clean = df_clean.rename(columns={"product_name": "name"})
```

```python
df_clean["review"] = df_clean["review_title"].fillna("") + " " + df_clean["review_content"].fillna("")
```

```python
df_clean = df_clean.drop(columns=["review_title", "review_content"])
```

```python
df_clean = df_clean[["name", "goods_or_services", "category", "about_product", "review"]]
```

```python
df_clean.head()
```

```python
air = pd.read_csv("Airline_review.csv")
air.head()
```

```python
cols_to_drop = [
    "Unnamed: 0",
    "Overall_Rating",
    "Review Date",
    "Verified",
    "Aircraft",
    "Type Of Traveller",
    "Seat Type",
    "Route",
    "Date Flown",
    "Seat Comfort",
    "Cabin Staff Service",
    "Food & Beverages",
    "Ground Service",
    "Inflight Entertainment",
    "Wifi & Connectivity",
    "Value For Money",
    "Recommended"
]

air = air.drop(columns=cols_to_drop)
```

```python
air = air.rename(columns={"Airline Name": "name"})
```

```python
# Etiqueta de clase
air["goods_or_services"] = "service"

# Categoría general
air["category"] = "Airline"

# No tenemos una descripción tipo "about product", así que la dejamos vacía
air["about_product"] = ""

# Fusionar título + texto de la review
air["review"] = air["Review_Title"].fillna("") + " " + air["Review"].fillna("")
```

```python
air = air.drop(columns=["Review_Title", "Review"])

air = air[["name", "goods_or_services", "category", "about_product", "review"]]
```

```python
air.head()
```

```python
air['about_product'] = "Commercial airline service offering domestic and international flights, including ticketing, boarding procedures, baggage services, cabin crew assistance, onboard amenities, seating comfort, and the overall passenger travel experience."
```

```python
# Unimos los dos datasets
df_merged = pd.concat([df_clean, air], ignore_index=True)

# Mostramos tamaño
print(df_merged.shape)

# Vemos primeras filas
df_merged.head()
```

```python
df_merged[df_merged["goods_or_services"] == "service"].head(1)
```

```python
service_names = df_merged[df_merged["goods_or_services"] == "service"]["name"].unique()
service_names
```

```python
import pandas as pd

# Cargar dataset
rest = pd.read_csv("Restaurant reviews.csv")

# Renombrar columnas
rest_clean = rest.rename(columns={
    "Restaurant": "name",
    "Review": "review"
})

# Eliminar columnas no necesarias
rest_clean = rest_clean.drop(columns=["Reviewer", "Rating", "Metadata", "Time", "Pictures"])

# Añadir columnas nuevas
rest_about = (
    "Commercial restaurant service offering dine in or takeaway meals, "
    "including food preparation, table service, menu variety, customer experience, "
    "staff interaction, and overall quality of the dining experience."
)

rest_clean["goods_or_services"] = "service"
rest_clean["category"] = "Restaurants"
rest_clean["about_product"] = rest_about

# Reordenar columnas
rest_clean = rest_clean[["name", "goods_or_services", "category", "about_product", "review"]]

rest_clean.head()
```

```python
df_merged = pd.concat([df_merged, rest_clean], ignore_index=True)
```
